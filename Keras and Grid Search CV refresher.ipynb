
# This notebook was done as practice in using Keras' Sequential neural network and applying Grid Search Cross-Validation on the neural network.

## The dataset used was given from an old advanced machine learning class. The dataset has 400 grayscale images that are 64x64 pixels. The dataset came with a csv file that contained the labels of each image.


```python

# importing libraries
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.wrappers.scikit_learn  import  KerasClassifier
from sklearn.model_selection  import  GridSearchCV
from sklearn.model_selection import train_test_split
from keras.utils import np_utils
import cv2
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import time

```

```python

# folder path to the image data used for this keras neural network
# there are 400 images, with 40 unique individuals
folder_path = r"C:\Users\KennoHead\Desktop\Data Science and Machine Learning Refresher\Face\\"

# processing the image data before using it for the keras neural network
# each face image is 64 x 64 pixels, all in JPG files and in grayscale
pixel_matrix = []

for i in range (0, 400, 1):
    
    current_image_data = cv2.imread(folder_path + str(i) + ".jpg", 0)
    pixel_matrix.append(current_image_data.flatten())


print("Process Done!")
```

```python

# reading the csv file that contains the labels for the images
# and merging the pixel data with the labels
pixels_labels = pd.read_csv(r"C:\Users\KennoHead\Desktop\Data Science and Machine Learning Refresher\Face\label.csv")
pixels_df = pd.DataFrame(pixel_matrix)
pixels_df = pd.concat([pixels_df, pixels_labels], axis = 1)

print("Process Done!")
```

```python

# setting up the training and testing sets

pixels_features_cols = list(pixels_df.columns)
pixels_features_cols.remove('Label')

pixels_features = pixels_df[pixels_features_cols]
pixels_labels = pixels_df['Label']


features_training, features_testing, labels_training, labels_testing = train_test_split(pixels_features, pixels_labels, test_size = .3, random_state = 0)

# preparing the data further before using the keras neural network
features_training /= 255
features_testing /= 255
labels_training = np_utils.to_categorical(labels_training, 40)
labels_testing = np_utils.to_categorical(labels_testing, 40)

print("Process Done!")
```

```python

# defining a model builder for the Keras Classifier
# note that input size is based on the number of pixels for each image
# and the output size for the last layer is to match the number of unique labels
def model_creator():
    
    model = Sequential()
    input_size = 4096
    hidden_neurons = 1024
    output_size = 40
    
    # adding a second layer: a hidden layer
    model.add(Dense(hidden_neurons, input_dim = input_size))
    model.add(Activation('sigmoid'))
    
    # adding a third layer: an output layer
    model.add(Dense(output_size, input_dim = input_size))
    model.add(Activation('softmax'))
    
    model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
    
    return model
    
    
seed = 0
np.random.seed(seed)

model = KerasClassifier(build_fn = model_creator, verbose = 2)

print("Process Done!")
```

```python

# dictionary for the parameter grid to test the batch sizes and the epochs
# cross validation value was picked to be 10
parameter_grid = {'batch_size':[30, 40, 50, 70, 100], 'epochs': [5, 10, 15, 20, 25, 30]}

start_time = time.perf_counter()

grid_search_cv = GridSearchCV(model, parameter_grid, cv = 10, scoring = 'neg_log_loss')
grid_search_cv.fit(features_training, labels_training)

end_time = time.perf_counter()

print(grid_search_cv.best_score_)
print(grid_search_cv.best_params_)

```

```python

# getting the results of the grid search cross validation
grid_search_cv_time_minutes = (end_time - start_time) / 60
best_score = grid_search_cv.best_score_
best_parameters = grid_search_cv.best_params_
best_model = grid_search_cv.best_estimator_.model

print('Grid Search CV time in minutes : ' + str(grid_search_cv_time_minutes) + ' minutes.')
print('Test Size: 30%, random state: 0.')
print('Best Score found in Grid Search CV ', '%0.4f' % best_score)
print('Best parameters found in Grid Search CV: ', best_parameters, '.')

label_predictions = best_model.predict(features_testing, verbose = 1)
score = best_model.evaluate(features_testing, labels_testing, verbose = 1)

print('The accuracy of the best model found in Grid Search CV is: ', '%0.4f' % score[1], '.')
```
