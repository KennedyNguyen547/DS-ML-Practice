
# This jupyter notebook is practice on using cross validation with k-nearest neighbor, decision tree, and logistic regression classifiers. The dataset used is the heart disease UCI dataset from kaggle.com



```python

import numpy as np

import pandas as pd

# reading the data
heart_df = pd.read_csv(r"C:\Users\KennoHead\Desktop\Data Science and Machine Learning Refresher\heart.csv")

# checking the dataframe
heart_df.head()
```

```python

from sklearn.model_selection import train_test_split

# preparing two kinds of training and testing splits
# one pair of trainning and testing sets using the original data
# the other pair using data with some features one hot encoded

heart_df_ohe = pd.get_dummies(heart_df, columns = ['sex', 'cp', 'thal'])

heart_df_features_cols = list(heart_df.columns)
heart_df_features_cols.remove('target')
heart_df_features = heart_df[heart_df_features_cols]
heart_df_labels = heart_df['target']


heart_df_ohe_features_cols = list(heart_df_ohe.columns)
heart_df_ohe_features_cols.remove('target')
heart_df_ohe_features = heart_df_ohe[heart_df_ohe_features_cols]
heart_df_ohe_labels = heart_df_ohe['target']

X_training, X_testing, Y_training, Y_testing = train_test_split(heart_df_features, heart_df_labels, test_size =.3, random_state = 0)

X2_training, X2_testing, Y2_training, Y2_testing = train_test_split(heart_df_ohe_features, heart_df_ohe_labels, test_size = .3, random_state = 0)
```

```python

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score

# using cross validation using accuracy with the original data first


k_value = 5
dec_tree_rs_value = 0
k_folds = 10
max_iter_steps = 1500

k5_knn_clssr = KNeighborsClassifier(n_neighbors = k_value)
dec_tree_rs = DecisionTreeClassifier(random_state = dec_tree_rs_value)
log_reg_clssr = LogisticRegression(max_iter = max_iter_steps)

k5_knn_cv_acc_list = cross_val_score(k5_knn_clssr, X_training, Y_training, cv = k_folds, scoring ='accuracy')
dec_tree_cv_acc_list = cross_val_score(dec_tree_rs, X_training, Y_training, cv = k_folds, scoring ='accuracy')
log_reg_clssr_cv_acc_list = cross_val_score(log_reg_clssr, X_training, Y_training, cv = k_folds, scoring ='accuracy')

# note that using cross_val_score will return a list of accuracy scores

k5_knn_mean_cv_acc = k5_knn_cv_acc_list.mean()
dec_tree_mean_cv_acc = dec_tree_cv_acc_list.mean()
log_reg_mean_cv_acc = log_reg_clssr_cv_acc_list.mean()


```

```python

# from the results, we can see that logistic regression gave the best results so far.
print(k5_knn_mean_cv_acc)
print(dec_tree_mean_cv_acc)
print(log_reg_mean_cv_acc)
```

```python

# now using data with some one hot encoding to repeat the steps above using cross validation
# we can simply copy the code from above and append a "2" to the identifiers 

k5_knn_clssr2 = KNeighborsClassifier(n_neighbors = k_value)
dec_tree_rs2 = DecisionTreeClassifier(random_state = dec_tree_rs_value)
log_reg_clssr2 = LogisticRegression(max_iter = max_iter_steps)

k5_knn_cv_acc_list2 = cross_val_score(k5_knn_clssr, X2_training, Y2_training, cv = k_folds, scoring ='accuracy')
dec_tree_cv_acc_list2 = cross_val_score(dec_tree_rs, X2_training, Y2_training, cv = k_folds, scoring ='accuracy')
log_reg_clssr_cv_acc_list2 = cross_val_score(log_reg_clssr, X2_training, Y2_training, cv = k_folds, scoring ='accuracy')

k5_knn_mean_cv_acc2 = k5_knn_cv_acc_list2.mean()
dec_tree_mean_cv_acc2 = dec_tree_cv_acc_list2.mean()
log_reg_mean_cv_acc2 = log_reg_clssr_cv_acc_list2.mean()
```

```python

# we can see from the results that with some one hot encoding
# the performance for knn did not improve
# the performance for decision tree gotten worse
# and the performance for logistic regression improved slightly

print(k5_knn_mean_cv_acc2)
print(dec_tree_mean_cv_acc2)
print(log_reg_mean_cv_acc2)
```
